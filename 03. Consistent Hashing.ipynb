# Consistent Hashing

Consistent Hashing is a technique used in distributed systems to distribute data across multiple servers efficiently.

It solves a major problem in normal hashing:

```text
When servers are added or removed,
huge amount of data gets remapped.
```

---

# Problem with Normal Hashing

Suppose:

```text
hash(key) % N
```

Where:
- key = user/session/cache key
- N = number of servers

---

# Example

3 servers:

```text
Server A
Server B
Server C
```

Formula:

```text
hash(key) % 3
```

---

# Issue

If one new server added:

```text
hash(key) % 4
```

Now almost ALL keys remap.

Example:

| Key | Old Server | New Server |
|---|---|---|
| user1 | A | D |
| user2 | B | A |
| user3 | C | B |

Huge cache misses happen.

---

# Goal of Consistent Hashing

When server changes:
- only small subset of keys should move

NOT all keys.

---

# Core Idea

Create a circular hash space called:

```text
Hash Ring
```

---

# Hash Ring Visualization

Imagine values from:

```text
0 → 2^32
```

arranged in a circle.

Example:

```text
        Server A
           |
           |
Server C ------- Server B
```

---

# Step 1 — Hash Servers

Each server gets a position on ring.

Example:

```text
hash(Server A) = 20
hash(Server B) = 50
hash(Server C) = 80
```

---

# Step 2 — Hash Keys

Hash the data key.

Example:

```text
hash(user123) = 60
```

---

# Step 3 — Clockwise Mapping

Move clockwise until first server found.

Example:

```text
Key = 60
Next server clockwise = Server C
```

So key stored in:

```text
Server C
```

---

# Important Property

If new server added:

```text
Only nearby keys move
```

NOT all keys.

---

# Virtual Nodes

## Problem Without Virtual Nodes

Suppose server positions become uneven.

```text
A ---------------- B -- C
```

Now:
- A handles huge traffic
- B/C handle less traffic

---

# Solution → Virtual Nodes

Each physical server gets multiple virtual positions.

Example:

```text
A1 A2 A3
B1 B2 B3
C1 C2 C3
```

Now load spreads evenly.

---

# Benefits of Virtual Nodes

| Benefit | Why |
|---|---|
| Better load balancing | Even distribution |
| Fault tolerance | Smoother redistribution |
| Scalability | Easier node addition/removal |

---

# Node Failure Scenario

Suppose:

```text
Server C crashes
```

Only keys owned by C move to next server.

Rest remain unchanged.

---

# Complexity

Lookup complexity:

```text
O(log N)
```

using:
- Sorted map
- BST
- TreeMap

---

# Real-World Usage

- Redis Cluster
- Memcached
- Apache Cassandra
- Load Balancers
- CDN systems

---

# Interview Architecture

```text
Clients
   ↓
Hash Function
   ↓
Hash Ring
   ↓
Target Server
```

---

# Advantages

| Advantage | Description |
|---|---|
| Minimal redistribution | Few keys move |
| Scalable | Easy node add/remove |
| Fault tolerant | Handles node failure |
| Efficient | Balanced distribution |

---

# Disadvantages

| Disadvantage | Solution |
|---|---|
| Uneven distribution | Virtual nodes |
| Complex implementation | Libraries/frameworks |

---

# Interview Summary

Consistent Hashing distributes keys across servers using a circular hash ring.

When servers are added or removed:
- only a small subset of keys get remapped

Unlike traditional modulo hashing where almost all keys move.

To improve load balancing:
- virtual nodes are used so that each physical server appears multiple times on the ring.
