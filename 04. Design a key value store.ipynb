
# Design a Key-Value Store

A Key-Value Store stores data in:

```text
key → value
```

Example:

```text
"user123" → "{name:'Shubham', age:25}"
```

Popular examples:
- Redis
- Amazon DynamoDB
- Apache Cassandra
- Memcached

---

# 1. Clarify Requirements

## Functional Requirements

- GET and PUT support
- DELETE support
- Persistence
- Replication
- Transactions
- Strong/Eventual consistency

---

## Non-Functional Requirements

- High availability
- Scalability
- Low latency
- Fault tolerance
- Durability

---

# 2. Basic APIs

```python
PUT(key, value)

GET(key)

DELETE(key)
```

# 3. High-Level Architecture

```text
Client
   ↓
Load Balancer
   ↓
API Servers
   ↓
Storage Nodes
```

---

# 4. Core Data Structure

Use:

```text
Hash Map
```

Because:
- O(1) average lookup
- Fast insertion
- Efficient retrieval

---

# 5. Single Machine Design

Data stored in memory:

```text
key → value
```

---

# Problem with Single Machine

Issues:
- memory limit
- SPOF
- scaling issues

Need distributed system.

---

# 6. Distributed Key-Value Store

Architecture:

```text
Clients
   ↓
Coordinator/API Layer
   ↓
Distributed Storage Nodes
```

---

# 7. Data Partitioning

Use Consistent Hashing.

Example:

```text
user123 → Server B
user456 → Server D
```

Benefits:
- minimal rebalancing
- scalable
- fault tolerant

---

# 8. Replication

```text
Primary Node
   ↓
Replica 1
Replica 2
```

Improves:
- availability
- durability

---

# 9. Replication Strategies

## Synchronous Replication

Pros:
- strong consistency

Cons:
- slower writes

---

## Asynchronous Replication

Pros:
- fast writes

Cons:
- temporary inconsistency

---

# 10. CAP Theorem

| Property | Meaning |
|---|---|
| Consistency | Same data everywhere |
| Availability | System always responds |
| Partition Tolerance | Survives network failures |

Most KV stores choose:
- AP or CP

---

# 11. Write Path

```text
1. Client sends PUT(key,value)
2. API server hashes key
3. Determine target node
4. Write to primary node
5. Replicate to replicas
6. Return success
```

---

# 12. Read Path

```text
1. Client sends GET(key)
2. Hash key
3. Find storage node
4. Retrieve value
5. Return response
```

---

# 13. Persistence

## Write Ahead Log (WAL)

---

## SSTables / LSM Trees

- Cassandra

# 21. Tradeoffs

| Choice | Benefit | Drawback |
|---|---|---|
| Strong consistency | Accurate data | Higher latency |
| Eventual consistency | Fast | Temporary stale reads |
| Replication | High availability | More storage |
| Caching | Fast reads | Cache invalidation |

# 23. Final Interview Summary

A distributed key-value store can be designed using:
- Consistent hashing for partitioning
- Replication for durability
- Hash maps for fast lookup
- WAL for persistence
- Distributed storage nodes for scalability

Depending on use case:
- choose strong consistency or eventual consistency.
