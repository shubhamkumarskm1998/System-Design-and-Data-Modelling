### OSI model 

explains how data travels from one computer to another over a network.

It has 7 layers, where each layer performs a specific task.

Think of it like sending a parcel:

- You write message
- Pack it
- Address it
- Transport company moves it
- Receiver opens and reads it

Networking works similarly.

# OSI Model (Open Systems Interconnection Model)

OSI model explains how data travels from one computer to another over a network.

It has **7 layers**, where each layer performs a specific task.

---

# 7 Layers of OSI Model

| Layer | Name | Main Work | Examples |
|---|---|---|---|
| 7 | Application | User interaction | HTTP, FTP, SMTP |
| 6 | Presentation | Encryption/Formatting | SSL, JPEG, UTF-8 |
| 5 | Session | Manages sessions/connections | NetBIOS |
| 4 | Transport | Reliable delivery | TCP, UDP |
| 3 | Network | Routing/IP addressing | IP, Router |
| 2 | Data Link | MAC addressing | Ethernet, Switch |
| 1 | Physical | Cables/signals | WiFi, Fiber, RJ45 |

---

# TCP vs UDP

## TCP
- Reliable
- Ordered packets
- Guarantees delivery

Used in:
- Websites
- Banking
- Emails

## UDP
- Faster
- Unreliable
- No guarantee of delivery

Used in:
- Gaming
- Streaming
- Video calls

# Practical Understanding

When you open a website:

```text
HTTP request
→ TCP
→ IP
→ Ethernet/WiFi
→ Cable/Signal
```

At receiver side it unwraps in reverse order.
